```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')
```

```python
# creating the dataset
data = {'Nano':103, 'Tiny':94, 'Small':71, 'Medium':31, 'Large':17}
models = list(data.keys())
fps = list(data.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(models, fps, color ='green',
        width = 0.4)

addlabels(models, fps)
 
plt.xlabel("Models")
plt.ylabel("FPS")
plt.title("FPS comparison in GTX 1060 6GB Notebook GPU")
plt.show()
```
