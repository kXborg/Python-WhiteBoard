```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import numpy as np
import cv2

fig = plt.figure(figsize=(20,20))
ax = fig.gca(projection='3d')   # Create the axes

# Data
X = np.linspace(-1, 1, 200)
Y = np.linspace(-1, 1, 200)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2

# Plot the 3d surface
surface = ax.plot_surface(X, Y, Z, cmap=cm.viridis, rstride = 1, cstride = 1)

# Set some labels
ax.set_xlabel('$W_1$', fontsize=30, rotation=180)
ax.set_ylabel('$W_2$', fontsize=30)
ax.set_zlabel('Loss (J)', fontsize=30)
ax.set_zticks([0, 1, 2])
ax.set_xticks([-1.0, -0.5, 0.0, 0.5, 1.0])
ax.set_yticks([-1.0, -0.5, 0.0, 0.5, 1.0])

# ax.xaxis._axinfo['juggled'] = (0,1,0)
# ax.yaxis._axinfo['juggled'] = (1,0,1)
# ax.zaxis._axinfo['juggled'] = (0,0,0)
plt.show()

fig.canvas.draw()
graph = cv2.cvtColor(np.asarray(fig.canvas.buffer_rgba()), cv2.COLOR_RGBA2BGR)
cv2.imwrite('graph.jpg', graph)
```
