```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
cv2.__version__
```

```python
img = cv2.imread('download.png')
plt.imshow(img[...,::-1])
```

```python
lb = np.array([0,50,50])
ub = np.array([10,255,255])
mask = cv2.inRange(img, lb, ub)
```

```python
plt.imshow(mask, cmap = 'gray')
```

```python
print(type(mask))
```

```python
for i in range(0, img.shape[0]):
    for j in range(0, img.shape[1]):
        if mask[i][j] > 0:
            print(mask[i][j])
```
