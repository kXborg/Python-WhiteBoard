```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set up the SimpleBlobdetector with default parameters.
params = cv2.SimpleBlobDetector_Params()

params.blobColor = 0

params.minDistBetweenBlobs = 2

# Filter by Area.
params.filterByArea = False

# Filter by Circularity
params.filterByCircularity = True
params.minCircularity = 0.8

# Filter by Convexity
params.filterByConvexity = True
params.minConvexity = 0.8

# Filter by Inertia
params.filterByInertia =True
params.minInertiaRatio = 0.8
# Create SimpleBlobDetector
detector = cv2.SimpleBlobDetector_create(params)
```

```python
image1 = cv2.imread('shapes-2.jpg',0)
keypoints1 = detector.detect(image1)
```

```python
canvas = cv2.cvtColor(image1, cv2.COLOR_GRAY2BGR)
```

```python
for keypt in keypoints1:
    x, y = keypt.pt
    dia = keypt.size
    x, y, dia = int(x), int(y), int(dia/2)
    print('Points: ', x, ',', y, '\n', 'Diameter : ', dia)
    out_img = cv2.circle(canvas, (x, y), dia, (0, 255, 0), 4)
```
