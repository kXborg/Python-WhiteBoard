```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
# x axis values
x = [1,2,3,4,5,6]
# corresponding y axis values
y = [2,4,1,5,2,6]
 
# plotting the points
plt.plot(x, y, color='green', linestyle='dashed', linewidth = 3,
         marker='o', markerfacecolor='blue', markersize=12)

# setting x and y axis range
plt.ylim(1,8)
plt.xlim(1,8)
 
# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')
 
# giving a title to my graph
plt.title('Some cool customizations!')
# function to show the plot
plt.show()

# redraw the canvas
fig.canvas.draw()

# convert canvas to image
img = cv2.cvtColor(np.asarray(fig.canvas.buffer_rgba()), cv2.COLOR_RGBA2BGR)
# img  = img.reshape(fig.canvas.get_width_height()[::-1] + (3,))

# img is rgb, convert to opencv's default bgr
img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
plt.imshow(img);
```

```python
# set width of bar
barWidth = 0.25
fig = plt.figure(figsize=(20,14))

# set height of bar
DNN = [94, 160, 290, 520, 868]
Pytorch = [93, 160, 315, 594, 947]
ONNX = [71, 129, 215, 372, 669]

# Set position of bar on X axis
br1 = np.arange(len(DNN))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

# Make the plot
bar1 = plt.bar(br1, DNN, color ='b', width = barWidth,
        edgecolor ='grey', label = "ONNX-DNN")
for bar in bar1:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + .04, yval)
    
bar2 = plt.bar(br2, Pytorch, color ='g', width = barWidth,
        edgecolor ='grey', label = "Ultralytics-Pytorch")
for bar in bar2:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + .04, yval)

bar3 = plt.bar(br3, ONNX, color ='r', width = barWidth,
        edgecolor ='grey', label = "Ultralytics-ONNX")
for bar in bar3:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + .04, yval)

# Adding Xticks
plt.xlabel('YOLOv5 models', fontweight ='bold', fontsize = 13)
plt.ylabel('Inference time CPU b1', fontweight ='bold', fontsize = 13)
plt.xticks([r + barWidth for r in range(len(DNN))],
        ['Nano', 'Small', 'Medium', 'Large', 'Extra large'])

plt.legend()
# plt.show()

# redraw the canvas
fig.canvas.draw()

# convert canvas to image
img = cv2.cvtColor(np.asarray(fig.canvas.buffer_rgba()), cv2.COLOR_RGBA2BGR)
# img  = img.reshape(fig.canvas.get_width_height()[::-1] + (3,))

# img is rgb, convert to opencv's default bgr
img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
```

```python
cv2.imwrite('yolov5-platform-comparisons.jpg', img)
```
