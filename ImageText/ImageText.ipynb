## Labelling Plan
There are many ways to achieve this. In our case, we will define a base image which is little bit bigger than the original image.

Let's make it 100 pixel taller than the actual image. The extra space will be used for adding text.

```python
# Define base container.
base_img = np.ones((720 + 100, 1280), dtype=np.uint8)
# Convert to BGR format.
base_img = cv2.cvtColor(base_img, cv2.COLOR_GRAY2BGR)

# Define text canvas.
text_overlay_region = 255*np.ones((100, 1280), dtype=np.uint8)
# Convert to BGR format.
text_overlay_region = cv2.cvtColor(text_overlay_region, cv2.COLOR_GRAY2BGR)
```

```python
plt.imshow(base_img, cmap='gray');
```

```python
plt.imshow(text_overlay_region, cmap='gray');
```

### Put Text

```python
# Text overlay.
cv2.putText(text_overlay_region, 'Example Text Here', (20, 70), 
           cv2.FONT_HERSHEY_DUPLEX, 2, (0,255,0), 2)

plt.imshow(text_overlay_region[...,::-1]);
```

### Adding text region

```python
base_img[0:100, 0:1280] = text_overlay_region
```

```python
plt.imshow(base_img[...,::-1]);
```

### Adding image

```python
base_img[100:820, 0:1280] = img
```

```python
plt.imshow(base_img[...,::-1]);
```
